#### 1. Import pandas library

```python
import pandas as pd
import numpy as np
```

#### 2. Import users table:

```python

users = pd.read_csv("datasets/users_table.csv")
users.head()
```

#### 3. Rename Id column to userId

```python
users = users.rename(columns={"Id":"userId"})
```

#### 4. Import posts table:

```python
posts = pd.read_csv("datasets/posts_table.csv")
posts.head()
```

#### 5. Rename Id column to postId and OwnerUserId to userId

```python
posts = posts.rename(columns={'Id':'postId','OwnerUserId':'userId' })
```

#### 6. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userId,ViewCount,CommentCount

```python

users_df = users[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
users_df.head()
posts_df = posts[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
posts_df.head()
```

#### 7. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
df_merged = pd.merge(posts_df, users_df, on="userId", how="left")
df_merged.head()
```

#### 8. How many missing values do you have in your merged dataframe? On which columns?

```python
df_merged.isna().sum()
```

#### 9. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
#Missing values related to post are filled as to avoid losing valid post data. and missing userId kept empty as it might be unknown user.
df_merged.fillna({"ViewCount": 0}, inplace=True)
df_merged.fillna({"Reputation": 0}, inplace=True)
df_merged.fillna({"Views": 0}, inplace=True)
df_merged.fillna({"UpVotes": 0}, inplace=True)
df_merged.fillna({"DownVotes": 0}, inplace=True)
```

#### 10. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
#Id columns and count columns

df_merged.info()

int_columns = ["userId", "ViewCount", "Reputation", "Views", "UpVotes", "DownVotes"]
df_merged[int_columns] = df_merged[int_columns].astype("Int64")
```
